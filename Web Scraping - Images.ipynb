```python
#import library
import requests 
import bs4
```

```python
# send request to get https://en.wikipedia.org/wiki/Lantern_Festival
res = requests.get("https://en.wikipedia.org/wiki/Lantern_Festival")
```

```python
# use the BeautifulSoup to convert the text formatting.
soup = bs4.BeautifulSoup(res.text, "lxml")
soup
```

```python
#select all image tag
img_info = soup.select('img')
img_info
len(img_info)
```

```python
all_img = []
for i in range(len(img_info)):
    all_img.append('https:'+ img_info[i]['src'])
all_img
```

```python
image_link = []
for i in range(len(img_info)):
    try:
        image_link.append(requests.get(all_img[i]))
    except:
        n = i+1
        print("Img" + str(n) + ", URL is error, this image will not be stored in the folder.")    
len(image_link)
```

```python
import os
folder = os.path.join('myImages')
os.makedirs(folder, exist_ok=True)

for i in range(len(image_link)):
    file_name = 'NewImage'+ str(i+1) +'.jpg'
    name = os.path.join(folder, file_name)
    f = open(name, 'wb')
    f.write(image_link[i].content)
f.close()

print("Done for images exporting")
```
